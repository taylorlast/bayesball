```python
import statsapi
import requests
import json
import ast
import pandas as pd
from bs4 import BeautifulSoup
```

```python
# MLB Leagues are 103 and 104
teams = statsapi.get('teams',{'ver':'v1','season':'2021','activeStatus':'Y','leagueIds':'103,104','fields':'teams,name,id,league'})['teams']
MLB_teams = [i for i in teams]

# Extract the IDs for each team
MLB_team_ids = [i['id'] for i in MLB_teams]
MLB_team_ids
```

```python
# MLB_team_ids_str = '{}'.format(MLB_team_ids)[1:-1]
pitchers = []
for team in MLB_team_ids:
    pitchers.append([i for i in statsapi.get('team_roster',{'ver':'v1','teamId':'{}'.format(team)})['roster'] if (i['position']['code'] == '1') & (i['status']['code'] == 'A')])
```

```python
pitcher_data = []
for i in pitchers:
    for j in i:
        pitcher_data.append(j['person'])
pitcher_df = pd.DataFrame(pitcher_data)
pitcher_df
```
